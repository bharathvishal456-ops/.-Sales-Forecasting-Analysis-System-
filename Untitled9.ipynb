```python
print('Checking for missing values:')
display(df.isnull().sum())
```

```python
# Convert 'Order Date' and 'Ship Date' to datetime objects
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')

print('\nData types after conversion:')
display(df[['Order Date', 'Ship Date']].dtypes)
```

```python
print('\nChecking for duplicate rows:')
duplicate_rows = df.duplicated().sum()
if duplicate_rows > 0:
    print(f'Found {duplicate_rows} duplicate rows. Removing duplicates...')
    df.drop_duplicates(inplace=True)
    print('Duplicates removed.')
else:
    print('No duplicate rows found.')
```

```python
total_sales = df['Sales'].sum()
average_sales = df['Sales'].mean()
max_sales = df['Sales'].max()
min_sales = df['Sales'].min()

print(f"Total Sales: {total_sales:,.2f}")
print(f"Average Sales: {average_sales:,.2f}")
print(f"Maximum Sales: {max_sales:,.2f}")
print(f"Minimum Sales: {min_sales:,.2f}")
```

```python
print('\nDescriptive statistics for Sales:')
display(df['Sales'].describe())
```

## 4. Sales Trend Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate sales by Order Date to analyze trends over time
daily_sales = df.groupby('Order Date')['Sales'].sum().reset_index()

plt.figure(figsize=(15, 7))
sns.lineplot(x='Order Date', y='Sales', data=daily_sales)
plt.title('Sales Trend Over Time')
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

## 5. Store/Product Analysis

```python
# Sales by Region
sales_by_region = df.groupby('Region')['Sales'].sum().sort_values(ascending=False).reset_index()
print('\nSales by Region:')
display(sales_by_region)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Sales', data=sales_by_region, palette='viridis')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Top 10 Best-Selling Products
products_sales = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10).reset_index()
print('\nTop 10 Best-Selling Products:')
display(products_sales)
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='Sales', y='Product Name', data=products_sales, palette='magma')
plt.title('Top 10 Best-Selling Products')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.show()
```

```python
# Top 10 Worst-Selling Products
products_sales_worst = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=True).head(10).reset_index()
print('\nTop 10 Worst-Selling Products:')
display(products_sales_worst)
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='Sales', y='Product Name', data=products_sales_worst, palette='cividis')
plt.title('Top 10 Worst-Selling Products')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.show()
```

```python
# Extract year, month, and day of week from Order Date
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month_name()
df['Order DayOfWeek'] = df['Order Date'].dt.day_name()

print('Extracted time components:')
display(df[['Order Date', 'Order Year', 'Order Month', 'Order DayOfWeek']].head())
```

## 8. Data Visualization

```python
# Sales Distribution (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales'], bins=50, kde=True)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Monthly Sales Trend
monthly_sales = df.groupby('Order Month')['Sales'].sum().reindex([    'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Order Month', y='Sales', data=monthly_sales, palette='cubehelix')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Yearly Sales Trend
yearly_sales = df.groupby('Order Year')['Sales'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Order Year', y='Sales', data=yearly_sales, palette='rocket')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```
